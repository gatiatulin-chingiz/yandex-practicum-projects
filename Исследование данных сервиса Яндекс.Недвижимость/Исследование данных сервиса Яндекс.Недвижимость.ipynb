## Проверка и замена NAN значений

Проверим датасет на наличие NAN значений в каждой колонке.

```python
df.isna().sum()
```

Определим процентное соотношение пропущенных данных в каждой колонке.

```python
df.isna().mean()*100
```

```python
def pass_value_barh(df):
    ((df.isna().mean()*100)
    .to_frame()
    .rename(columns = {0:'space'})
    .query('space > 0')
    .sort_values(by = 'space', ascending = True)
    .plot(kind= 'barh', figsize=(19,6), legend = False, fontsize = 16)
    .set_title('Пример' + "\n", fontsize = 22, color = 'SteelBlue'))
```

```python
pass_value_barh(df)
```

Видно, что в колонках *`ceiling_height`, `is_appartment`, `balcony`, `parks_nearest`, `ponds_nearest`* находится самый большой процент пропущенных значений. Это может сильно искажать результаты исследований.

Заменим значения NAN по столбцам, пряменяя к кажому столбцу свой подход:

I. Создадим словарь с ключами в виде колонок датасета, и значениями в виде необходимых манипуляций с данными.
1. Известно, что количество аппартаментов по сравнению с количеством квартир очень мало, поэтому заменим все пропущенные значения на значение *False* (не аппартаменты) и тем самым приравняем всю выборку к типу "квартира".
2. Пропущенные значения в колонке *`balcony`* с количество балконов возможно возникли из-за отсутствия в таких квартирах балконов вовсе. Поэтому пропущенные значения заменим на нулевые значения.
3. Количество парков и прудов, аналогично количеству балконов, может рядом и не быть.

II. Передадим словарь с ключами и необходимыми функциями в функцию *fillna* для обработки и выведем результат.

```python
# Обработка NAN значений
value = {
    'is_apartment': False,
    'balcony': 0,
    'parks_around3000': 0,
    'ponds_around3000': 0,
}

df = df.fillna(value=value)
df.isna().mean()*100
```

Необходимые замены произведены. Замена в колонках *`airports_nearest` и `parks_nearest`* из-за невозмоности их замены числовым значением (0 будет означать, что квартира находится в парке или в аэроротре), а удаление приведет к большой потере данных.

Проведем удаление значений NAN в колонках *`floors_total`, `locality_name`* т.к. их значение в выборке не велико.

```python
df = df.dropna(subset=['floors_total', 'locality_name'])
```

```python
df.isna().mean()*100
```

### Категоризация этажности

Для возможности анализа объявления в зависимости от этажности проведем категоризацию этажности объектов недвижимости.
Категоризация объектов проведем в соответствии со следующей логикой:
1. Объекты недвижимости на 1 этаже определим в категорию *первый*.
2. Объекты недвижимости на последнем этаже в категорию *последний*.
3. Объекты недвижимости на остальных этажах в категорию *другой*.

```python
# Фукнция категоризации этажности объектов недвижимости
try:
    def floor_group(row):
        if row['floor'] == 1:
            return 'первый'
        elif row['floor'] == row['floors_total']:
            return 'последний'
        return 'другой'
    # Применение функции floor_group к выборке по столбцу floor_type
    df['floor_type'] = df.apply(floor_group, axis=1)
except:
    print('Произошла ошибка')
```

Теперь составим сводную таблицу по цене за квадратный метр в выбранных населенных пунктах.

```python
ten_locality_sort.reset_index().plot(kind='barh', x='locality_name',  y='area_price')
```

Выведем максимальную и минимальную цену за квадратный метр в данных населенных пунктах.

```python
# Вывод максимальной цены за квадратный метр
(df.query('locality_name in @ten_locality.index')
   .pivot_table(values='area_price', index='locality_name')
   .sort_values(by='area_price', ascending=False)
   .head(1)
)
```

```python
# Вывод минимальной цены за квадратный метр
(df.query('locality_name in @ten_locality.index')
   .pivot_table(values='area_price', index='locality_name')
   .sort_values(by='area_price', ascending=False)
   .tail(1)
)
```

Из проведенного анализа можно сделать вывод, что наибольшая средняя стоимость квардатного метра жилья приходится на г. Санкт-Петербург. Вторым по стоимости квадратного метра является г. Пушкин.
```python
(df.groupby('locality_name')['area_price']
   .agg({'count', 'mean'})
   .sort_values(by = 'count',
                ascending = False)
   .head(10)
   .sort_values(by='mean', ascending=False)
   .style.format("{:,.0f}"))
```
